```python
import chromadb 
from pprint import pprint

chroma_client = chromadb.Client()

collection = chroma_client.get_or_create_collection(name="documents")

docs = [
  {
    "id": "doc1",
    "content": "Fruits are natural foods that come from plants. They are rich in vitamins, minerals, and fibres. Apples, bananas, and oranges are some popular fruits that contribute to a healthy diet."
  },
  {
    "id": "doc2",
    "content": "School subjects are the various areas of study in an educational system. Common subjects include mathematics, physics, biology, history, and languages. Each subject helps students develop different skills and knowledge."
  },
  {
    "id": "doc3",
    "content": "Python is a high‑level programming language known for its simplicity and readability. It is widely used in web development, data analysis, artificial intelligence, and automation."
  },
  {
    "id": "doc4",
    "content": "GitHub is a platform for hosting and collaborating on software projects. It uses Git for version control and allows developers to share code, track changes, and contribute to open‑source projects."
  },
  {
    "id": "doc5",
    "content": "Lycée d’Excellence de Rabat (Lydex) is a Moroccan high‑school that offers two distinct academic paths: a rigorous academic track (Tronc Commun scientifique, option française) and a sport‑studies track (Sport‑Études) that combines high‑level sports training in disciplines like athletics, judo, taekwondo, swimming and golf with academic excellence. :contentReference[oaicite:1]{index=1}"
  }
]


collection.upsert(
    ids=[doc['id'] for doc in docs],
    documents=[doc['content'] for doc in docs]
)


result1 = collection.query(
    query_texts=['repository'],
    n_results=1
)
pprint(result1) #* Doc that related to Github

result2 = collection.query(
    query_texts="Hi my name is Ahmed Bouramdane And I am a student at lydex high school",
    n_results=1
)
pprint(result2) #* doc related to Lydex 
```
